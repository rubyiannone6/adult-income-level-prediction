## replacing outliers

```python
def outliers(x, f):
  q25 = x.quantile(.25)
  q75 = x.quantile(.75)
  IQR = q75-q25
  l = q25 - IQR*f
  u = q75 + IQR*f
  return l, u

l, u = outliers(df['age'],1.5)
print(f'Lower Boundary= {l}, Upper Boundary={u}')
```

```python
upperbound = df['age'] > u
upperbound.sum()
```

```python
for i in df[upperbound].index:
  df.loc[i,'age'] = u
df['age'].plot.box(color='red',title='',whis=1.5,showmeans='True')
```

```python
l, u = outliers(df['hours-per-week'],1.5)
print(f'Lower Boundary= {l}, Upper Boundary={u}')
```

```python
upperbound = df['hours-per-week'] > u
upperbound.sum()
```

```python
lowerbound = df['hours-per-week'] < l
lowerbound.sum()
```

```python
for i in df[upperbound].index:
  df.loc[i,'hours-per-week'] = u
for i in df[lowerbound].index:
  df.loc[i,'hours-per-week'] = l
df['hours-per-week'].plot.box(color='blue',title='',whis=1.5,showmeans='True')
```
